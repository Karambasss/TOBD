5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты без отзывов должны отсутствовать в данной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного `DataFrame`, содержащую информацию об этом отзыве.

```python
recipes
```

```python
reviews
```

```python
new_table = pd.merge(
    recipes,
    reviews,
    how='inner',
    left_on='id',
    right_on='recipe_id'
)[['id', 'name', 'user_id', 'rating']]
new_table
```

```python
unic_recipes = set(recipes["id"]) 
unic_review = set(reviews["recipe_id"])
rec_without_rev = pd.Series(list(unic_recipes - unic_review)) # получим айди тех recipe, на которых не оставили отзыва
print(rec_without_rev)
test_recipe = rec_without_rev.sample(1) # возвратит 1 строку (индекс и сам айди)
print("------------------------------------------------------------------------")
print(test_recipe)
sample_recipe = test_recipe.values[0]

#так как нету id recipe у тех кто в review не оставил отзыва на recipe, а мы создали таблицу, состояшую только из тех, кто написал отзыв, то докажем, пройдя по всем значениям id из таблицы new recipes и сравнивая с рандомным id user который не оставил отзыв -> тем самым, такой строки в таблице не найдется, ибо в таблице только те кто написал отзыв
new_table[new_table["id"] == sample_recipe]
```
