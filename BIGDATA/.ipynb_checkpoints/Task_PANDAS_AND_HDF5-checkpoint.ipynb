```python
import pandas as pd
path = "exam_datasets/ТОБД21_датасеты/датасеты_к_заданию_2/sp500hst.txt"
```

```python
my_table2 = pd.read_csv(path,names = ['date', 'ticker', 'open', 'high', 'low', 'close', 'volume'],header=None)
my_table2
```

```python
my_table2["date"] = pd.to_datetime(my_table2["date"], format="%Y%m%d")
```

```python
my_table2
```

```python
dates = my_table2["date"].unique()

tickers = my_table2["ticker"].unique()
```

```python
my_table = pd.DataFrame(columns=tickers, index=dates)
my_table
```

```python
for x in dates:
    my_table.loc[x] = my_table2[my_table2["date"] == x].groupby("ticker")["volume"].sum()
```

 вопрос В массиве чисел, хранящихся в файле random.hdf5, найти строку (вывести её индекс и содержащиеся значения), в которой более всего значений предыщающих среднее значение по всему массиву. Выполнить задание с использованием dask.array, распараллелив процесс обработки данных (использование Dask должно приводить к истинной параллелльной обработке данных).

```python
import h5py
import dask.array as da

p = "exam_datasets/ТОБД21_датасеты/датасеты_к_заданию_3/random.hdf5"

with h5py.File(p, "r") as h5:
    dasa = da.array(h5["data_set_1"][:])
    
dasa[(dasa > dasa.mean()).sum(axis=1).argmax()].compute() #посчитали сумму и индекс самого большого элемента
```

```python
t = [2,3,5,56,7,8]

r = da.from_array(t)
r[r.argmax()].compute()
```
