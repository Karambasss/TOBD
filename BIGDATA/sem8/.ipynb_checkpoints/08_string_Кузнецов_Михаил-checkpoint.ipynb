1.2 Напишите функцию `show_info`, которая для рецепта по его `id` создает строку (в смысле объекта python) с описанием следующего вида:

```
"Название"

1. Шаг 1.
2. Шаг 2.
----------
#тэг1 #тэг2
```

    
Данные для создания строки получите из файлов `recipes_sample_with_tags_ingredients.csv`, `steps_sample.xml` (__ЛР4__) и `tags_sample.csv` (__ЛР5__). 
Выведите созданную строку на экран.

````python
def read_xml(path: str) -> BeautifulSoup:
    with open(path, encoding='utf-8') as f:
        return BeautifulSoup(''.join(f.readlines()), 'lxml')
    
steps_sample_xml = read_xml(DATA_INPUT_DIR + 'steps_sample.xml')
````

````python
#перебрать все степы и запихнуть их в одну строку, используюя sep как разделитель шагов

sep2 = '$'
steps_dict_2 = {'id': [], 'steps': []}

for recipe in steps_sample_xml.find_all('recipe'):
#     print(recipe.find('id'))
    steps_dict_2['id'].append(int(recipe.find('id').text))
    step_str = ''
    for step in recipe.find_all('step'):
        step_str += sep2
        step_str += step.text
    steps_dict_2['steps'].append(step_str)
steps_df_2 = pd.DataFrame(steps_dict_2)
steps_df_2.head()
````

````python
def select_columns(data_frame, column_names):
    return data_frame.loc[:, column_names]


recipes_steps_df = recipes.merge(steps_df_2).set_index('id')

columns = ['name', 'tags', 'steps']
recipe_id_name_tags_name = select_columns(recipes_steps_df, columns)
recipe_id_name_tags_name
# recipes_steps_df.drop(['minutes', 'contributor_id','submitted', 'n_steps', 'n_tags', 'ingredients', 'n_ingredients', 'description'], axis=1)
````

````python
def show_info(id_recipe, df, sep1):
    
    recipe = df.loc[id_recipe]
    
    good_name = "\"" + recipe["name"][0].upper() + recipe["name"][1:] + "\""
    
    name = "{:^100}".format(good_name)

    steps_in_string = []
    
    step_numb = 1
    
    steps_2 = recipe['steps'].split(sep1)[1:]  #отрезаем разделитель
    
    #print(steps_2)
    
    for step in steps_2:
        
        #step = step[0].upper() 
        
        stepU = step[0].upper() + step[1:]

        if step[-1] != "?" or step[-1] != "!" or step[-1] != ".":
            steps_in_string.append(f'{step_numb}. {stepU}.')
        else:
            steps_in_string.append(f'{step_numb}. {stepU}')
        step_numb += 1
    
    #print(steps_in_string)
    
    steps = '\n'.join(steps_in_string)
    
    #print(steps)

    sep_before_tag = '-' * 100
    
    tags = []
    
    for t in recipe['tags'].split(';'):
        tags.append('#' + t)
    
    #print(tags)
        
    tags = ''.join(tags)
    
    good_arr = [name, '\n', steps, sep_before_tag, tags]
    
    #print(good_arr)
    
    return '\n'.join(good_arr)

info = show_info(44123,recipe_id_name_tags_name, "$")
print(info)
````

2.2 Найдите все смайлики в отзывах. Смайлик состоит из трех частей: глаза (символ `=` или `:`), нос (символ `-`), губы (символ `)` или `(`). Смайлик может иметь вид "глаза-нос-губы" или "губы-нос-глаза". Нос может отсутствовать.

````python
pattern_smile = re.compile(r'([=:]-{0,1}[()])|([()]-{0,1}[=:])')
count = 0 
for x in review_list:
    a = pattern_smile.findall(str(x))
    if len(a) != 0:
        count+=1
        print(a)
        
count
````

2.3 Проверьте, что все даты в датасете имеют вид "YYYY-MM-DD". Продемонстрируйте работу вашего решения, приведя пример из датасета и контрпример не из датасета.

````python
date_list = reviews_2["date"].to_list()
date_list
````

````python
pattern_date = re.compile(r'[0-9]{4}-([0][1-9]|[1][0-9])-([0][1-9]|[12][0-9]|[3][0-1])')
count_data = 0

for x in date_list:
    if pattern_date.search(x) != None:
        count_data+=1
    else:
        print("Дата не подходит!")

print("1: Продемонстрирована работа регулярки, так как не вывелось ни одного сообщения о том, что дата не подходит!"'\n')        

print(count_data)
print(reviews_2["date"].shape[0])        
print("2: Продемонстрирована работа регулярки, так как все кол-во наших дат из исходной таблицы совпало с количеством отобранных дат с помощью регулярки")
````

````python
bad_date = ['3333,4444,5555', '333=dada=31','fafasf-flafl-fafks','1999-01-00']
for x in bad_date:
    if pattern_date.search(x) == None:
        print("Дата не подходит!")
    else:
        print("Дата подходит!")
````

3.2 Посчитайте кол-во уникальных слов в датасете (без учета регистра).

````python
arr = []
for review in reviews.to_list():
    arr.extend(nltk.word_tokenize(str(review).lower()))
arr
````

````python
p = re.compile(r'[\W^ ]')

for x in arr:
    if bool(p.match(x)): 
        arr.remove(x)
arr
````

````python
nltk.FreqDist(arr) #сколько встретилось слов в тексте
````

````python
nltk.FreqDist(arr).N() #количество слов в тексте
````

````python
nltk.FreqDist(arr).B() #количество уникальных слов в тектсе
````
