```python
import pandas as pd
```

```python
usa_table = pd.read_csv("data/us-counties.csv", sep = ",", parse_dates=["date"])
usa_table
```

```python
dates = usa_table["date"]
dates
```

```python
usa_2020 = usa_table[(dates >= "2020-01-01") & (dates <= "2020-12-31")]
usa_2020
```

```python
states = [x for x in pd.unique(usa_2020["state"])]
states
```

```python
month = [x for x in pd.unique(usa_2020["date"].dt.month)]
month 
```

```python
df = pd.DataFrame(columns = month)
df
```

```python
for x in states:
    month_grouped_states = usa_2020[usa_2020["state"] == x]["deaths"].groupby(usa_2020["date"].dt.month).sum()
    print(month_grouped_states)
    #loc обращается к индексу элемента (если не нашел,то создаст новый)
    df.loc[x] = month_grouped_states
df = df.fillna(0)
```
