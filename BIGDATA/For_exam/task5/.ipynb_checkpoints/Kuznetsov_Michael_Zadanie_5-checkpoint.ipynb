По данным из файла data/meals.json сформировать словарь,
в котором по идентификатору блюда можно получить список ингредиентов.

```python
import json
def read_json(path: str):
    with open(path) as file:
        return json.load(file)
```

```python
meals_book = read_json('data/meals.json')
meals_book
```

```python
def find_ingredients_via_id(meals_id):
    for x in meals_book.get("meals"):
        if x["idMeal"] == str(meals_id):
            return [x[f"strIngredient{i}"] for i in range(1,21)]
```

```python
find_ingredients_via_id(53049)
```
