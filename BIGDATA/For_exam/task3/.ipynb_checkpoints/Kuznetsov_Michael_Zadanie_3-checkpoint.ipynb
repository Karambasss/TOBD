Датасет: album_artist.xlsx
С помощью кода на Python с использованием xlwings решить задачу. Вынести названия артистов на отдельный лист "Артисты" и присвоить каждому артисту уникальный идентификатор. Заменить названия артистов на исходном листе на идентификаторы с листа "Артисты".


```python
import xlwings as xw
import numpy as np
import pandas as pd
from xlwings.constants import AutoFillType
from xlwings import constants
```

```python
DATA_DIR = "data/"
```

```python
#подключаемся к существующему эксель файлу
my_exel = xw.Book(DATA_DIR+"album_artist.xlsx")
```

```python
# получаем лист книги Excel:
sheet1 = my_exel.sheets['Sheet1']
```

```python
#добавляем лист после sheet1
artist_sheet = my_exel.sheets.add("Артисты", after=sheet1)
```

```python
artist_sheet.range("A1").value = "IdArtist"
artist_sheet.range("B1").value = "Name"
```

```python
#верный способ

artists = sheet1.range("D2:D348").options(np.array).value
artists
```

```python
artists_unique = np.unique(artists)
artists_unique
```

```python
artist_sheet.range("B2").options(transpose=True).value = artists_unique
```

```python
id_artist = [i for i in range(1, artists_unique.shape[0]+1)]
id_artist
```

```python
artist_sheet.range("A2").options(transpose=True).value = id_artist
```

```python
artist_sheet.range("C1").options(transpose=True).value = artist_sheet.range("A1:A205").value
```

```python
artist_sheet.range("A1").options(transpose=True).value = artist_sheet.range("B1:B205").value
```

```python
artist_sheet.range("B1").options(transpose=True).value = artist_sheet.range("C1:C205").value
```

```python
artist_sheet.range("C1:C205").delete()
```

```python
sheet1.range("E2").formula = '=VLOOKUP(D2,Артисты!$A$2:$B$205,2,FALSE)'
sheet1.range('E2').api.AutoFill(sheet1.range('E2:E348').api, AutoFillType.xlFillDefault)
```

```python
sheet1.range("D2").options(transpose=True).value = sheet1.range("E2:E348").value
```

```python
sheet1.range("E2:E348").delete()
```

```python
my_exel.save()
```
