1. Напишите функцию, которая считывает файл формата xml из каталога `reviewers_full` и по данным этого файла формирует список словарей, содержащих следующие ключи: `id`, `username`, `name`, `sex`, `country`, `mail`, `registered`, `birthdate`, `name_prefix`, `country_code`. Часть из этих значений в исходном файле хранится в виде тэгов, часть - в виде атрибутов тэгов. Для конкретного человека какие-то из этих ключей могут отсутствовать. 



```python
INPUT_DIR = 'reviewers_full/'
```

```python
def my_func(filename):
    f_xml = open(INPUT_DIR + filename, 'r')
    data = BeautifulSoup(f_xml, 'xml')
    users = []

    for user in data.find_all('user'):
        user_dict = dict()
        
        if user.id is not None:
            user_dict['id'] = int(user.id.string)
        if user.username is not None:
            user_dict['username'] = user.username.string
        if user.sex is not None:
            user_dict['sex'] = user.sex.string
        if user.country is not None:
            user_dict['country'] = user.country.string
        if user.mail is not None:
            user_dict['mail'] = user.mail.string
        if user.registered is not None:
            user_dict['registered'] = user.registered.string
        
        if user.birthdate is not None:
             user_dict['birthdate'] = user.birthdate.string
        
        if user.has_attr("prefix"):
            user_dict['name_prefix'] = user.attrs['prefix']
        
        if user.country is not None and user.country.has_attr('code'):
            user_dict['country_code'] = user.country.attrs['code']
        users.append(user_dict)

    f_xml.close()
    return users
```

3. Задекорируйте функцию из задания 1 при помощи `dask.delayed` и создайте список `reviewers`, состоящий из 5 объектов `delayed` (по одному объекту на файл). Из списка объектов `delayed`, создайте `dask.bag` при помощи метода `db.from_delayed`. Добавьте ключ `birth_year`, в котором хранится год рождения человека. Оставьте в выборке только тех людей, которые __наверняка__ моложе 1980 года. Преобразуйте поле `id` к целому типу.

```python
from datetime import datetime

@delayed
def my_func_dec(filename):
    f_xml = open(INPUT_DIR + filename, 'r')
    data = BeautifulSoup(f_xml, 'xml')
    users = []

    for user in data.find_all('user'):
        user_dict = dict()
    
        if user.id is not None:
            user_dict['id'] = int(user.id.string)
        if user.username is not None:
            user_dict['username'] = user.username.string
        if user.sex is not None:
            user_dict['sex'] = user.sex.string
        if user.country is not None:
            user_dict['country'] = user.country.string
        if user.mail is not None:
            user_dict['mail'] = user.mail.string
        if user.registered is not None:
            user_dict['registered'] = user.registered.string
        
        if user.birthdate is not None:
            user_dict['birthdate'] = user.birthdate.string
            user_dict['birth_year'] = datetime.fromisoformat(user.birthdate.string).year
        else:
            user_dict['birth_year'] = None
        if user.has_attr("prefix"):
            user_dict['name_prefix'] = user.attrs['prefix']
        
        if user.country is not None and user.country.has_attr('code'):
            user_dict['country_code'] = user.country.attrs['code']
        users.append(user_dict)

    f_xml.close()
    return users
```
