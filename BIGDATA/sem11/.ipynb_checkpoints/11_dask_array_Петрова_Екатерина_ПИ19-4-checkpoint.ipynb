6. Найдите максимальную продолжительность рецепта. Ограничьте максимальную продолжительность рецептов сверху значением, равному 75% квантилю.

```python
def limitRecipes(dask_array):
    q_result = dask_array.to_dask_dataframe().quantile(0.75).compute()
    dask_array[dask_array > q_result] = q_result
    return dask_array
```

```python
new_recipes = recipes[:, 1]
print("Максимальная продолжительность рецепта:", da.max(new_recipes).compute() )
new_recipes = limitRecipes(new_recipes)
new_recipes
```

```python
new_recipes.compute()
```

```python
print("Максимальная продолжительность рецепта после ограничения:", da.max(new_recipes).compute())
```

8. Работая с исходным файлом в формате `hdf5`, реализуйте алгоритм подсчета среднего значения в блочной форме и вычислите с его помощью среднее значение второго столбца в массиве.

Блочный алгоритм вычислений состоит из двух частей:
1. загрузка фрагмента за фрагментом данных по `blocksize` элементов и проведение вычислений на этим фрагментом;
2. агрегация результатов вычислений на различных фрагментах для получения результата на уровне всего набора данных.

Важно: при работе с `h5py` в память загружаются не все элементы, а только те, которые запрашиваются в данный момент

```python
sums = []

lengths = []

blocksize = 100000

data_s = h5py.File('./data/minutes_n_ingredients_full.hdf5', "r")["recipe"][:,1]

arr_len = data_s.shape[0]

for i in range(0, arr_len, blocksize):
    chunk = data_s[i: i + blocksize]
    sums.append(chunk.sum())
    lengths.append(len(chunk))
    
total = sum(sums)
lengths = sum(lengths)
print("Ответ: ", total/lengths)
```
